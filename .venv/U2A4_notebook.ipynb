## 1. ¿Qué son los splits de un dataset?

Un **split** (partición) consiste en dividir el dataset original en subconjuntos separados antes de entrenar un modelo de inteligencia artificial. Los dos más comunes son:

| Subconjunto | Proporción típica | Uso |
|---|---|---|
| **Train set** | 70% – 80% | El modelo aprende patrones con estos datos |
| **Test set** | 20% – 30% | Se evalúa el modelo con datos que nunca vio |

### ¿Por qué son importantes?

Si entrenamos y evaluamos con los **mismos datos**, el modelo puede simplemente memorizar las respuestas y obtener resultados artificialmente perfectos. Este problema se conoce como **sobreajuste (overfitting)**.

Al separar los datos:
- Medimos el desempeño **real** del modelo ante datos nuevos.
- Detectamos si el modelo generaliza bien o solo memoriza.
- Garantizamos una evaluación **objetiva e imparcial**.

### ¿Qué es la semilla fija (random_state)?

La división aleatoria puede producir resultados distintos cada vez que se ejecuta. Fijar una **semilla** (por ejemplo `random_state=42`) garantiza que siempre se genere exactamente la misma partición, haciendo el experimento **reproducible**: cualquier persona con el mismo código y dataset obtendrá los mismos archivos `train.csv` y `test.csv`.

### ¿Qué es stratify?

El parámetro `stratify=y` asegura que la proporción de cada clase (calidad del vino) se mantenga igual en train y test. Así evitamos que, por azar, una clase quede subrepresentada en algún subconjunto.

## 2. Importación de librerías

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import os

print('Librerías importadas correctamente ✓')
print(f'  pandas  : {pd.__version__}')
print(f'  numpy   : {np.__version__}')
```

## 3. Carga del dataset

Cargamos el archivo **winequality-red.csv** desde la misma carpeta donde está este notebook.

```python
# Cargar el dataset desde archivo local
df = pd.read_csv('winequality-red.csv', sep=';')

# Renombrar columnas para eliminar espacios
df.columns = [
    'fixed_acidity', 'volatile_acidity', 'citric_acid',
    'residual_sugar', 'chlorides', 'free_sulfur_dioxide',
    'total_sulfur_dioxide', 'density', 'pH',
    'sulphates', 'alcohol', 'quality'
]

print('Dataset cargado correctamente ✓')
print(f'Filas    : {df.shape[0]}')
print(f'Columnas : {df.shape[1]}')
df.head()
```

## 4. Exploración básica del dataset

```python
# Información general
print('=== Información del dataset ===')
df.info()
print()

# Verificar valores nulos
print('=== Valores nulos por columna ===')
print(df.isnull().sum())
print()

# Distribución de la variable objetivo
print('=== Distribución de calidad (quality) ===')
print(df['quality'].value_counts().sort_index())
```

## 5. Partición train/test con semilla fija

```python
# Separar características (X) y etiqueta (y)
X = df.drop('quality', axis=1)
y = df['quality']

# Parámetros de la partición
TEST_SIZE   = 0.2
RANDOM_SEED = 42

# Realizar la partición
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=TEST_SIZE,
    random_state=RANDOM_SEED,
    stratify=y
)

# Reconstruir DataFrames completos
train_df = X_train.copy()
train_df['quality'] = y_train.values

test_df = X_test.copy()
test_df['quality'] = y_test.values

# Resultados
print('=== Resultado de la partición ===')
print(f'Total original  : {len(df):>6} filas (100%)')
print(f'Train set       : {len(train_df):>6} filas ({len(train_df)/len(df)*100:.0f}%)')
print(f'Test set        : {len(test_df):>6} filas ({len(test_df)/len(df)*100:.0f}%)')
print()
print(f'Semilla utilizada (random_state): {RANDOM_SEED}')
print(f'Criterio de división: estratificado por calidad (stratify=y)')
```

```python
# Verificar que la distribución de clases se mantuvo proporcional
print('=== Proporción de calidad en cada split ===')
print()
print('Original:')
print((df['quality'].value_counts(normalize=True)*100).round(1).sort_index().to_string())
print()
print('Train:')
print((train_df['quality'].value_counts(normalize=True)*100).round(1).sort_index().to_string())
print()
print('Test:')
print((test_df['quality'].value_counts(normalize=True)*100).round(1).sort_index().to_string())
```

## 6. Guardar archivos en la carpeta /splits

```python
# Crear carpeta splits/ si no existe
os.makedirs('splits', exist_ok=True)

# Guardar archivos CSV
train_df.to_csv('splits/train.csv', index=False)
test_df.to_csv('splits/test.csv',   index=False)

# Verificar que se guardaron correctamente
print('Archivos guardados en la carpeta splits/ ✓')
print()
for archivo in ['splits/train.csv', 'splits/test.csv']:
    verificacion = pd.read_csv(archivo)
    print(f'  {archivo}')
    print(f'    Filas    : {len(verificacion)}')
    print(f'    Columnas : {list(verificacion.columns)}')
    print()
```

## 7. Nota técnica

---

### ¿Qué son los splits y su importancia?

Los splits son particiones del dataset original en subconjuntos independientes. Su principal objetivo es separar los datos de **entrenamiento** (con los que el modelo aprende) de los datos de **prueba** (con los que se evalúa su desempeño real). Esta separación es fundamental para detectar problemas como el sobreajuste y para garantizar que la métrica de evaluación refleje la capacidad real del modelo de generalizar ante datos nuevos que nunca ha visto.
